Q1. How has honey production yield changed from 1998 to 2021?


```python
q1 = df.groupby('year')[['yieldpercol']].sum().reset_index().sort_values(by='year')
type(q1)
```

```python
q1
```

```python
plt.figure(figsize=(10,6.5))
sns.barplot(q1['year'],q1['yieldpercol'],edgecolor='black',palette='crest')
plt.title('Total Honey Yields across USA per Year',size=15)
plt.xlabel('Years',size=12)
plt.ylabel('Yields',size=12)
plt.xticks(rotation=90)
plt.savefig('Total Honey Yields across USA per Year.jpg')
plt.show()

# plt.figure(figsize=(10,6.5))
# sns.lineplot(q1['year'],q1['yieldpercol'])
```

### 1. How has honey production yield changed from 1998 to 2021?<br>

### Ans 1 = We can clearly see that honey yields have been dropping continously. Though during some years yields increased in comparison to previous year, but altogether it has dropped throughout the years. In 1998 total yields in USA were 3008 which has dropped to 1868 in the year 2021.

Q3. Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006, which was when concern over Colony Collapse Disorder spread nationwide?


```python
q3 = df[df['year']<=2006]
q3
```

```python
q3_1 = q3.groupby('year')[['numcol','yieldpercol']].sum().reset_index()
q3_1
```

```python
2637000 - 2621000.0
```

```python
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(8,10),sharex=True)   #nrows = 1 and ncols = 2
#ax1 : name of plot 1
#ax2 : name of plot 2
#multi_index
ax1.bar(q3_1['year'],q3_1['numcol'],color='blue',edgecolor='black')
ax1.set_title('Sum of Colonies Year-wise')
ax1.set_ylabel('Colony')
ax1.set_xlabel('Year')


ax2.plot(q3_1['year'],q3_1['yieldpercol'],color='purple',lw=3,marker='*',markersize=8)
ax2.set_title('Sum of Yield Per Colony Year-wise')
ax2.set_ylabel('Yield Per Colony')
ax2.set_xlabel('Year')
ax2.grid()

plt.show()
```

Q4. Are there any patterns that can be observed between total honey production and value of production every year?

```python
df.head().sort_values(by='totalprod',ascending=False)
```

```python
q4 = df.groupby('year')[['totalprod','prodvalue']].sum()
df_a=df[['totalprod','prodvalue']]//1000000
```

```python
q4.sort_values(by='totalprod',ascending=False)
```

```python
plt.figure(figsize=(8,5))
ax = sns.scatterplot(data=df_a,x='totalprod',y='prodvalue')
ax.set_ylabel('Product Value(in Million $)')
ax.set_xlabel('Total Production(in Million Pounds)')
ax.set_title('Total Honey Production vs Value of Honey Production')
plt.grid()
plt.show()
```
